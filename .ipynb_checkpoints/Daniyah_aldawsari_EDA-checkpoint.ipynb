```python
dict_ = pd.Series(df_alexa.homepage_domain.values,index = df_alexa.username).to_dict()
dict_
for key, value in dict_.items():
    dict_[key] = value.replace('http://www.','').replace( 'https://www.','').replace('https://', '').replace('http://', '').replace('/','')
    if key.lower() != key:
        dict_[key.lower()] = dict_[key]
        del dict_[key]
dict_
```

```python
# Ensure Irrelevant tweets are removed (these are probably 
# retweets but will consider them irrelavant since there is
# no way of telling there source)
remove = []
for i,x in df_twitter.iterrows():
    if x['username'] in dict_:
        pass
    else:
        remove.append(x['username'])
```

```python
# Group twitter data per day
twitter_pro_df = df_twitter.groupby(['date','username'])
twitter_final_df = twitter_pro_df.apply(lambda x: pd.Series({'total_tweets' : x.count()[0],
                                  'likes' : x['likes'].sum(), 
                                  'links': x.tweet.str.count(dict_[x['username'].head(1).item().lower()]).sum(),
                                  'hashtags_count': x.tweet.str.count('#').sum(),
                                  'hashtags_total': x.hashtags.sum(),
                                  'timezone': x.timezone.iloc[0],
                                  'mentions_sum': x.mentions.sum(),
                                   'word_count': x.tweet.count(),
                                   'link_to_tweet':x.link.sum(),
                                  'replies': x['replies'].sum() , 'retweets': x['retweets'].sum()} ))
```

Our data is hevely right skewed we will explor a couple of normalizing options and choose what fits best:
- log
- log1p
- quantile transformer


```python

from sklearn.preprocessing import QuantileTransformer, quantile_transform
log_transform = np.log(df['reach_per_million'].values.reshape(-1,1))
log1p_transform = np.log1p(df['reach_per_million'].values.reshape(-1,1))
quantile_transform = quantile_transform(df['reach_per_million'].values.reshape(-1,1),output_distribution='normal', random_state=0).squeeze()
```

```python
f, axes = plt.subplots(nrows=2, ncols=2)
plt.figure(figsize=(500,500))
axes[0,0].hist(df['reach_per_million'].values.reshape(-1,1))
#ax0.set_xlim([0, 2000])
axes[0,0].set_ylabel('Probability')
axes[0,0].set_xlabel('Target')
axes[0,0].set_title('Target distribution')

axes[0,1].hist(log_transform)
axes[0,1].set_ylabel('Probability')
axes[0,1].set_xlabel('Target')
axes[0,1].set_title('Transformed target distribution')

axes[1,0].hist(log1p_transform)
axes[1,0].set_ylabel('Probability')
axes[1,0].set_xlabel('Target')
axes[1,0].set_title('Transformed target distribution')

axes[1,1].hist(quantile_transform)
axes[1,1].set_ylabel('Probability')
axes[1,1].set_xlabel('Target')
axes[1,1].set_title('Transformed target distribution')

#f.suptitle("Synthetic data", y=0.035)
#f.tight_layout(rect=[0.05, 0.05, 0.95, 0.95])

#X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```
