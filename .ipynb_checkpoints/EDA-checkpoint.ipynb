## Part 1: Setup the dataset

The first step was gathering the data itself. which included:
- tweet scrapping using twint library (file in path `gathering/scrapper/twitter.py`) used to produce all data in `data/twitter` folder
- twitter followers count using tweepy library and twitter api (file in path `gathering/api/twitter_followers.py`) used to produce file `data/twitter_followers.json`
- traffic data for one month from similarweb (file in path `gathering/api/similarweb_data.ipynb`) to produce `data/alexa.json`
- website info data from amazon alexa using myawis (file in path `gathering/api/alexa_url_data.ipynb`) used to produce `data/url_info.json`
- crunchbase open data file (file in path `data/organizations.csv`)
- mapper file used to map twitter_url with valid account types `twitter_map.json`
- websites categories from webshrinker stored in `data/url_categories.json` (file in path `gathering/api/categories_data.py`)

```python
import os, json
import pandas as pd
import glob
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
% matplotlib inline
from sklearn import preprocessing
from sklearn.preprocessing import scale
import scipy
import warnings
warnings.filterwarnings('ignore')
```

### Prepare alexa data

```python
# Crunchbase data
df_organizaitons = pd.read_csv('data/organizations.csv')
# Twitter mapper
df_mapper = pd.read_json('data/twitter_map.json')
# Traffic alt
df_traffic_alt = pd.read_csv('data/similarweb.csv')
# Links in count (parameters are set to prevent wrong datatype convertion in median load time)
df_url_info = pd.read_json('data/url_info.json', convert_dates=False, convert_axes=False)
# Categorizationgs 
df_url_categories = pd.read_json('data/url_categories.json')
```

```python
df_similar_merge = df_traffic_alt.merge(df_organizaitons, left_on='website', right_on='homepage_domain')
df_similar_merge = df_similar_merge.merge(df_url_info, left_on='homepage_url', right_on='website')
```

```python
# ensure url are all lowercase
df_similar_merge['twitter_url'] = df_similar_merge['twitter_url'].str.lower()
df_mapper['twitter_url'] = df_mapper['twitter_url'].str.lower()
df_mapper['username'] = df_mapper['username'].str.lower()
```

```python
df_similar_merge = df_similar_merge.merge(df_mapper , on='twitter_url', how='left')
```

```python
df_similar_merge.info()
```

### Prepare twitter data

```python
# Read data from twitter folder
df_twitter = pd.concat([pd.read_json(f,lines=True) for f in glob.glob('data/twitter/*.json')], ignore_index = True)
```

```python
df_followers = pd.read_json('data/twitter_followers.json')
```

```python
df_followers['username'] = df_followers['username'].str.lower()
```

```python
# Ensure twitter username is lowercase
df_twitter['username'] = df_twitter['username'].str.lower()
```

```python
df_twitter = df_twitter.merge(df_mapper, on='username')
```

```python
df_twitter.info()
```

```python
# Create dicionary used to find if user linked their website
dict_ = pd.Series(df_similar_merge.homepage_domain.values,index = df_similar_merge.username).to_dict()
dict_
for key, value in dict_.items():
    dict_[key] = value
    if key.lower() != key:
        dict_[key.lower()] = dict_[key]
        del dict_[key]
```

```python
# Remove twitter data for site that had no traffic data
df_twitter = df_twitter[df_twitter['username'].isin(df_similar_merge.username)]
```

```python
# Create twitter dataset with records representing data from each day
# Group twitter data per day
twitter_pro_df = df_twitter.groupby(['date','username'])
twitter_final_df = twitter_pro_df.apply(lambda x: pd.Series({'total_tweets' : x.count()[0],
                                  'likes' : x['likes'].sum(), 
                                  'links': x.tweet.str.count(dict_[x['username'].head(1).item().lower()]).sum(),
                                  'hashtags_count': x.tweet.str.count('#').sum(),
                                  'hashtags_total': x.hashtags.sum(),
                                  'timezone': x.timezone.iloc[0],
                                  'mentions_sum': x.mentions.sum(),
                                   'mentions_count': x.tweet.str.count('@').sum(),
                                   'word_count': x.tweet.str.split(' ').str.len().sum(),
                                   'link_to_tweet':x.link.sum(),
                                  'replies': x['replies'].sum() , 'retweets': x['retweets'].sum()} ))
```

```python
twitter_final_df.head()
```

### Create the dataset

```python
df_similar_merge = df_similar_merge.merge(df_url_categories, how='left', left_on='homepage_url', right_on='homepage_url')
```

```python
# Convert date string to datetime object
df_similar_merge['date'] = df_similar_merge['date'].astype('datetime64[ns]')
```

```python
# Merge all the data
df = twitter_final_df.merge(df_similar_merge,left_index=True,  right_on=['date','username'])
```

### Convert columns into approriate formats

```python
# Number of links linking to this side
df['links_in_count'] = pd.to_numeric(df['links_in_count'])
# Median load in seconds
df['median_load_time'] = pd.to_numeric(df['median_load_time'])
# percent of known sites that are slower
df['speed_percentile'] = pd.to_numeric(df['speed_percentile'])
```

```python
# Add is_weekend colum (assumption weekend in all arabic world country weeken is on friday and saturday)
df["is_weekend"] = ((df["date"].dt.dayofweek == 5) | (df["date"].dt.dayofweek == 4)).astype(float)
```

```python
# Null values in views
len(df[df['views'] == 0])
```

```python
# Replace 0 with nan to find a way to fill those missing values 
df.views.replace(0,np.NaN, inplace=True)
```

```python
# Null values in reach_per_million
len(df[df['views'].isna()])
```

### Since we have 258 rows we wont just drop these values we will do a forward fill backward fill with apply funciton and groupby to not fill the traffic data for the same site on different dates

```python
df['views'] = df.groupby(['username'], sort=False)['views'].apply(lambda x: x.ffill().bfill())
```

```python
df[df['views'].isna()].username.value_counts()
```

```python
# These values dont have any available traffic data so we will drop them
df.views.dropna(inplace=True)
```

```python
df['views'].isna().sum()
```

```python
df.dropna(subset=['views'], inplace=True)
```

### Transform the predicted variable

```python
sns.distplot(y)
print(y.skew())
print(scipy.stats.kurtosis(y, axis=0, fisher=True, bias=True))
```

### Our data is hevely right skewed we will explore a couple of normalizing options and choose what fits best:
- log
- log1p
- quantile transformer

```python
from sklearn.preprocessing import QuantileTransformer, quantile_transform
```

```python
log_transform = np.log(relations_no_dummy['views'].values.reshape(-1,1))
log1p_transform = np.log1p(relations_no_dummy['views'].values.reshape(-1,1))
quantile_transform = quantile_transform(relations_no_dummy['views'].values.reshape(-1,1),output_distribution='normal', random_state=0).squeeze()
```

```python
fig, ax =plt.subplots(2,2,figsize=(25,15))

sns.distplot(log_transform ,  ax=ax[0,0])
ax[0,0].set_title(('log transform skew: %s kurtosis: %s' %(scipy.stats.skew(log_transform),scipy.stats.kurtosis(log_transform, axis=0, fisher=True, bias=True))))
ax[0,1].set_title(('log1p transform skew: %s kurtosis: %s' %(scipy.stats.skew(log1p_transform),scipy.stats.kurtosis(log1p_transform, axis=0, fisher=True, bias=True))))

ax[1,1].set_title(('Original data skew: %s kurtosis: %s' %(scipy.stats.skew(y),scipy.stats.kurtosis(y, axis=0, fisher=True, bias=True))))

ax[1,0].set_title(('quantile transform skew: %s kurtosis: %s' %(scipy.stats.skew(quantile_transform),scipy.stats.kurtosis(quantile_transform, axis=0, fisher=True, bias=True))))

sns.distplot(log1p_transform,  ax=ax[0,1])
sns.distplot(quantile_transform,  ax=ax[1,0])
sns.distplot(y.values.reshape(-1,1), ax=ax[1,1])
fig.show()
```

- The quantile_transform gives an almost perfect result but since it impacts the linear relation we won't select it. 
- both the log and log1p give good results to prevent having negative values will go with the log1p


```python
relations_no_dummy['log1p_transform'] = log1p_transform
```
