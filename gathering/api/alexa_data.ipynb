### Extract Traffic Data using Alexa AWIS

```python
from myawis import *
from lxml import etree
import json
import pandas as pd
```

```python
# Read crunchbase organization list
df = pd.read_csv('../../data/organizations.csv')
```

```python
Access_Key_ID='***************' # Removed for security reasons
Secret_Access_Key='*************'  # Removed for security reasons
```

```python
# Check for Duplication
df[df.duplicated(keep=False, subset='homepage_url')]
```

```python
# Remove last duplicate
df.drop_duplicates(keep='last',subset='homepage_url',inplace=True)
```

```python
# Remove companies with no twitter url
df = df[df.twitter_url.notnull()]
```

```python
# Create username column for merging later on
df['username'] = df['twitter_url'].apply(lambda x: x.split('/')[-1])
```

```python
# Make it lowercase
df['username'] = df['username'].str.lower()
```

```python
# Call Alexa web information service and retrieve xml file
# Convert Xml to json list
obj = CallAwis(Access_Key_ID, Secret_Access_Key)
traffic_data = []
for company in df['homepage_url']:
    print('parsing '+company)
    twitter_url_var = df[df['homepage_url'] == company]['username'].item()
    month = 20180801
    trafficinfo = obj.traffichistory(company,myrange = 31,start = month)
    web = trafficinfo.findAll('aws:Site')[0]
    for data_element in trafficinfo.findAll('aws:Data'):
        data_attrs = dict(data_element.attrs)
        date = data_element.find('aws:Date')
        reach = data_element.find('aws:Reach')
        rank = data_element.find('aws:Rank')
        page_views = data_element.find('aws:PageViews')
        page_view_per_million = page_views.find('aws:PerMillion')
        page_view_per_user = page_views.find('aws:PerUser')
        reach_per_million = reach.find('aws:PerMillion')
        data_variable = {'website': web.text ,'date' : date.text, 'reach_per_million' : reach_per_million.text,
                            'rank': rank.text, 'page_views_per_million': page_view_per_million.text,
                            'page_view_per_user': page_view_per_user.text, 'twitter_user': twitter_url_var}
        traffic_data.append(data_variable)
traffic_data
```

```python
# Store list in json file
with open('../../data/alexa.json', 'w') as outfile:
    json.dump(traffic_data, outfile)
```
