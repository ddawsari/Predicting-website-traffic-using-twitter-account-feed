### Extract Url Data using Alexa AWIS

```python
from myawis import *
from lxml import etree
import json
import numpy as np
import pandas as pd
```

```python
# Read crunchbase organization list
df = pd.read_csv('../../data/organizations.csv')
```

```python
Access_Key_ID='***************' # Removed for security reasons
Secret_Access_Key='*************'  # Removed for security reasons
```

```python
# Check for Duplication
df[df.duplicated(keep=False, subset='homepage_url')]
```

```python
# Remove last duplicate
df.drop_duplicates(keep='last',subset='homepage_url',inplace=True)
```

```python
# Remove companies with no twitter url
df = df[df.twitter_url.notnull()]
```

```python
# Call Alexa web information service and retrieve xml file
# Convert Xml to json list
obj = CallAwis(Access_Key_ID, Secret_Access_Key)
list_linkin = []
for company in df['homepage_url']:
    print('parsing '+ company)
    urlinfo = obj.urlinfo(company)
    speed = np.nan
    links_in_count = np.nan
    median_load_time = np.nan
    percentile = np.nan
    try:
        speed = urlinfo.find('aws:Speed')
    except:
        pass
    try:
        links_in_count = urlinfo.find('aws:LinksInCount').text
    except:
        pass
    try:
        median_load_time = speed.find('aws:MedianLoadTime').text
    except:
        pass
    try:
        percentile = speed.find('aws:Percentile').text
    except:
        pass
            
    categories = []
    try:
        categories_element = urlinfo.find('aws:Categories')
        for category in categories_element.findAll('aws:CategoryData'):
            title = category.find('aws:Title').text
            categories.append(title)
    except:
        pass
    data_variable = {'website': company ,'links_in_count' : links_in_count, 'median_load_time' : median_load_time,
                     'speed_percentile': percentile, 'categories':categories}
    list_linkin.append(data_variable)
list_linkin
```

```python
# Store list in json file
with open('../../data/url_info.json', 'w') as outfile:
    json.dump(list_linkin, outfile)
```
